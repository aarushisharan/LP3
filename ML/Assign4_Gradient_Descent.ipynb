```python
start_x=2
precision=0.00001
rate=0.01
step_size=1
df=lambda x: 2*(x+3)
max_iterations=10000
iterations=0
gd=[]
```

```python
while(step_size>precision and iterations<max_iterations):
    prev=start_x
    start_x=start_x-rate*df(prev)
    step_size=abs(prev-start_x)
    iterations+=1
    gd.append(start_x)
    print(f"Iterations: {iterations}, Value: {start_x}")
```
